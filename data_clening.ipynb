```python
import re
def convert_to_gb(size_str):

    match = re.match(r'(\d+\.?\d*)\s*(GB|TB)?', size_str)
    if not match:
        return None



    value = float(match.group(1))
    unit = match.group(2) or 'GB'


    if unit == 'TB':
        return value * 1024
    elif unit == 'GB':
        return value
    else:
        return value

lap['Disk_Size_GB'] = lap['Disk_size'].apply(convert_to_gb)

```

```python
usd_to_inr = 83.00  # Example exchange rate


def convert_to_inr(price_str):

    price_str = price_str.replace('$', '').replace(',', '')

    price_usd = float(price_str)
    price_inr = price_usd * usd_to_inr
    return price_inr

lap['Price_INR'] = lap['Price'].apply(convert_to_inr)
```

```python
def categorize_cpu_model(text):
    text = text.lower()

    if 'i9' in text:
        return 'i9'
    elif 'i7' in text:
        return 'i7'
    elif 'i5' in text:
        return 'i5'
    elif 'ryzen' in text:
        return 'Ryzen'
    elif 'celeron' in text:
        return 'Celeron'
    elif 'i3' in text:
        return 'i3'
    else:
        return 'Other'

# Apply the categorization function
lap['Cpu_Type'] = lap['CPU_Model'].apply(categorize_cpu_model)
```
